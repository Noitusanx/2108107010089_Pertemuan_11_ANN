```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Importing the dataset
df = pd.read_csv('dataset/Health_insurance.csv')
df.head()
```

```python
# feature encoding
from sklearn.preprocessing import LabelEncoder
LE = LabelEncoder()
df['region'] = LE.fit_transform(df['region'])
print(LE.classes_)
print(np.sort(df['region'].unique()))
print('')
```

```python
# Split the data and classes
X = df[['age']].values
y = df['charges'].values
```

```python
# features scaling
from sklearn.preprocessing import StandardScaler
scaler_X = StandardScaler()
scaler_y = StandardScaler()
X_scaled = scaler_X.fit_transform(X)
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size=0.25, random_state=0)
```

```python
# Mengimpor library Keras dan turunannya
# Sekarang menggunakan TensorFlow 2.0
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.metrics import MeanSquaredError
```

```python
# Inisialisasi ANN
MesinRegresi = Sequential()

# Menambah input layer dan hidden layer pertama
MesinRegresi.add(Dense(units=6, kernel_initializer='uniform', activation='relu', input_dim=1))

# Menambah hidden layer kedua
MesinRegresi.add(Dense(units=6, kernel_initializer='uniform', activation='relu'))

# Menambah output layer
MesinRegresi.add(Dense(units=1, kernel_initializer='uniform', activation='linear'))

# Meenjalankan ANN
MesinRegresi.compile(optimizer='adam', loss='mean_squared_error', metrics=[MeanSquaredError()])


# Fitting ANN ke training set
MesinRegresi.fit(X_train, y_train, batch_size=10, epochs=100)

# Memprediksi hasil test set
y_pred = MesinRegresi.predict(X_test)
```
